### Tarea 3.5

Suponga que usted es un ayudante de un laboratorio de física. En un experimento se pide medir la posición de una masa que cae en un plano inclinado a intervalos regulares en el tiempo. El equipo disponible puede medir las posiciones con una precisión de $4\,\text{mm}$. Uno de los estudiantes usa los siguientes datos (en $\text{cm}$):

```python
datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]
```

Grafique estos datos con sus barras de error. Explique por qué el estudiante merece una nota de 1.0. Compare con los siguientes datos obtenidos por un estudiante que merece 7.0.

```python
datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]
```

## Desarrollo

```python
import matplotlib.pyplot as plt
import numpy as np
```

Usando la función "errorbar" de pyplot graficaremos los datos de ambos alumnos, junto con
las barras de error correspondientes.

```python
plt.errorbar(np.arange(0,8),datos,yerr=0.4,ecolor="red")
plt.title("Gráfica alumno nota 1")
plt.xlabel("Tiempo")
plt.ylabel("Posición(cm)")
plt.show()
```

```python
plt.errorbar(np.arange(0,8),datos_7,yerr=0.4,ecolor="red")
plt.title("Gráfica alumno nota 7")
plt.xlabel("Tiempo")
plt.ylabel("Posición(cm)")
plt.show()
```

Viendo ambas gráficas, notamos que los datos del primer alumno son demasiado perfectos para
un aparato de medición con un márgen de error así de grande, sus datos forman una parábola casi perfecta, por lo tanto se podría concluir entregó datos "cocinados". Por otro lado, el alumno de 7 tiene datos 
los cuales son más esperables de mediciones con error, mostrando una curva un poco más irregular.
